## 1.Importation et lecture du dataset


```python
import pandas as pd
train = pd.read_csv('C:/Users/hp/Downloads/train.csv')
print (train.head())
```

## 2. Le nombre de mots dans chaque Tweet : 


```python
train['word_count'] = train['tweet'].apply(lambda x: len(str(x).split(" ")))
train[['tweet','word_count']].head()
```

## 3. Le nombre de caractère de chaque tweet


```python
train['nbre_caractères'] = train['tweet'].apply(lambda x: len(str(x)))
train[['tweet','nbre_caractères']].head()
```

## 4. Nombre de Stopwords

```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

train['stopword_count'] = train['tweet'].apply(lambda x: len([word for word in str(x).lower().split() if word in stop_words]))
train[['tweet', 'stopword_count']].head()
```

## 5. Nombre de Hashtags

```python
import re

train['hashtag_count'] = train['tweet'].apply(lambda x: len(re.findall(r'\#\w+', x)))
train[['tweet', 'hashtag_count']].head()
```

## 6. Nombre de Chiffres

```python
train['numeric_count'] = train['tweet'].apply(lambda x: len([char for char in str(x) if char.isdigit()]))
train[['tweet', 'numeric_count']].head(20)
```

## 7. Nombre de mots majuscules ( Upper Case ) :


```python
train['uppercase_count'] = train['tweet'].apply(lambda x: len([word for word in str(x).split() if word.isupper()]))
train[['tweet', 'uppercase_count']].head()
```

## 8.	Transfomation des Tweets en Miniscule ( Lower Case ) : 

```python
train['tweet_lower'] = train['tweet'].apply(lambda x: str(x).lower())
train[['tweet', 'tweet_lower']].head()
```

## 9.	Elimination de la ponctuation : 


```python
train['tweet_nopunc'] = train['tweet_lower'].apply(lambda x: re.sub('[^\w\s]', '', str(x)))
train[['tweet', 'tweet_nopunc']].head()
```

## 10. Affichage des StopWords

```python
import nltk
from nltk.corpus import stopwords

# Display the list of stop words
stop_words = set(stopwords.words('english'))
print(stop_words)

# Remove certain stop words while keeping others
stop_words_remove = set(['not', 'no', 'don', "don't", 'aren', "aren't", 'couldn', "couldn't",
                         'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn',
                         "hasn't", 'haven', "haven't", 'isn', "isn't", 'mightn', "mightn't",
                         'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn',
                         "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't",
                         'wouldn', "wouldn't"])
stop_words_filtered = stop_words - stop_words_remove

# Apply stop word removal to the tweets
train['tweet_nostop'] = train['tweet_nopunc'].apply(lambda x: ' '.join([word for word in str(x).split() if word.lower() not in stop_words_filtered]))
train[['tweet', 'tweet_nopunc', 'tweet_nostop']].head()
```

## 11.	Affichage des mots rarement utilisé dans les tweets et les supprimer s'ils ne sont pas nécessaires.

```python
# Import the necessary libraries
from collections import Counter

# Count the frequency of each word in the tweets
word_count = Counter(' '.join(train['tweet_nostop']).split())

# Display the list of words with frequency = 1
rare_words = set([word for word, count in word_count.items() if count == 1])
print(rare_words)

# Remove the rare words from the tweets
train['tweet_clean'] = train['tweet_nostop'].apply(lambda x: ' '.join([word for word in str(x).split() if word not in rare_words]))
train[['tweet', 'tweet_clean']].head()
```

## 13.	tokenisation : diviser le texte en une séquence de mots puis les convertir en une série de mots.

```python
# Import the necessary libraries
import nltk


# Define a function to tokenize a given tweet
def tokenize_tweet(tweet):
    """
    Given a tweet, return the tweet tokenized into a sequence of words using the NLTK word_tokenize() function.
    """
    # Tokenize the tweet using the NLTK word_tokenize() function
    tokenized_tweet = nltk.word_tokenize(tweet)
    
    return tokenized_tweet

# Apply the tokenize_tweet function to the 'tweet_corrected' column of the DataFrame
train['tokens'] = train['tweet_corrected'].head(40).apply(tokenize_tweet)

# Display the first few rows of the DataFrame, including the original 'tweet' column, the version of the tweets with spelling errors corrected ('tweet_corrected'), and the tokenized version of the tweets ('tokens')
train[['tweet', 'tweet_corrected', 'tokens']].head(40)
```

## 14.	Sentiment Analysis : le score en fonction du sentiment exprimé.

```python
from textblob import TextBlob

# Define a function to perform sentiment analysis on a given tweet
def get_sentiment_score(tweet):
    """
    Given a tweet, return a score between -1 and +1 based on the sentiment expressed in the tweet.
    """
    # Create a TextBlob object from the tweet
    blob = TextBlob(tweet)
    
    # Use the TextBlob sentiment property to get the sentiment polarity (-1 to +1)
    sentiment_score = blob.sentiment.polarity
    
    return sentiment_score
train['sentiment_score'] = train['tweet_corrected'].head(40).apply(get_sentiment_score)
```

## 15. Histogramme des sentiments

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load the dataset
train = pd.read_csv('C:/Users/hp/Downloads/tweets.csv')

# Define a function to get the sentiment polarity of a given tweet
def get_tweet_sentiment(tweet):
    """
    Given a tweet, return the sentiment polarity of the tweet using the TextBlob library.
    """
    # Create a TextBlob object from the tweet
    blob = TextBlob(tweet)
    
    # Use the TextBlob sentiment property to get the sentiment polarity
    sentiment = blob.sentiment.polarity
    
    return sentiment

# Apply the get_tweet_sentiment function to the 'tweet' column of the DataFrame to get the sentiment polarity for each tweet
train['sentiment'] = train['tweet'].head(40).apply(get_tweet_sentiment)

# Create a histogram of the sentiment scores using matplotlib
plt.hist(train['sentiment'], bins=20)

# Add labels and title to the plot
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.title('Sentiment Analysis of Tweets')

# Show the plot
plt.show()
```
